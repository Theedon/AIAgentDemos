```python

from azure.storage.blob import (
    BlobServiceClient,
)

blob_service_client = BlobServiceClient.from_connection_string(
    os.environ["AZURE_STORAGE_CONNECTION_STRING"]
)

def upload_artifact_to_azure(
    file_data: bytes, file_name: str, container: str = "support-app-artifacts"
) -> str:
    """
    Uploads a file to an Azure Blob Storage container.

    Args:
        file_data (bytes): The file data to be uploaded.
        file_name (str): The name of the file to be uploaded.
        container (str, optional): The name of the Azure Blob Storage container. Defaults to "support-app-artifacts".

    Returns:
        str: The URL of the uploaded file.

    Raises:
        Exception: If there is an error during the upload process.
    """
    try:
        container_client = blob_service_client.get_container_client(container)

        # Ensure container exists
        try:
            container_client.get_container_properties()
        except Exception:
            container_client.create_container(public_access="container")

        # Upload the blob
        blob_client = container_client.get_blob_client(file_name)
        blob_client.upload_blob(file_data, overwrite=True)
        uploaded_file_url = blob_client.url
        return uploaded_file_url

    except Exception as e:
        print(f"Error during Azure upload: {e}")
        raise  # Re-raise exception
```

```python
import ast
import os
import re
import tempfile
import traceback
import uuid
from datetime import datetime
from typing import Optional, Type
from passlib.context import CryptContext

import matplotlib.pyplot as plt
from langchain.agents import (
    AgentExecutor,
    create_tool_calling_agent,
)
from langchain.tools import BaseTool
from langchain_core.prompts import ChatPromptTemplate
from pydantic import BaseModel, Field



GENERATED_ARTIFACTS_DIRECTORY = "generated_artifacts"
pwd_context = CryptContext(schemes=["bcrypt"], deprecated="auto")


def parse_and_execute_code(code: str) -> dict:
    """
    Parse the given Python code, execute it, and return the result along with a local namespace.

    Args:
        code (str): The Python code to execute.

    Returns:
        dict: Contains the execution result and local namespace.
    """

    print(
        "**********LLM GENERATED CODE***********\n{}\n**************************************".format(
            code
        )
    )
    try:
        tree = ast.parse(code)
        last_node = tree.body[-1] if tree.body else None

        # Capture the last expression's result
        if isinstance(last_node, ast.Expr):
            target = ast.Name(id="_result", ctx=ast.Store())
            assign = ast.Assign(targets=[target], value=last_node.value)
            tree.body[-1] = ast.fix_missing_locations(assign)

        local_namespace = {}
        global_namespace = {
            "__builtins__": __builtins__,
            # Packages for DOCX processing
            "docx": __import__("docx"),
            "Document": __import__("docx").Document,
            # Packages for CSV processing
            "pandas": __import__("pandas"),
            "numpy": __import__("numpy"),
            # Packages for Excel processing
            "openpyxl": __import__("openpyxl"),
            "pyexcel": __import__("pyexcel"),
            "pyxlsb": __import__("pyxlsb"),
            # Packages for PDF processing
            "PyPDF2": __import__("PyPDF2"),
            "pypdf": __import__("pypdf"),
            "pymupdf": __import__("fitz"),  # `fitz` is the pymupdf module
            "pdfplumber": __import__("pdfplumber"),
            "reportlab": __import__("reportlab"),
            "camelot": __import__("camelot"),
            "tabula": __import__("tabula"),
            "PyPDFium2": __import__("pypdfium2"),
            # Packages for charting
            "matplotlib": __import__("matplotlib"),
            "pyplot": __import__("matplotlib").pyplot,
        }
        compiled_code = compile(tree, filename="<ast>", mode="exec")
        exec(compiled_code, global_namespace, local_namespace)
        execution_result = {
            "result": local_namespace.get("_result", "no result returned"),
            "namespace": local_namespace,
        }
        return execution_result

    except Exception as e:
        error_details = traceback.format_exc()
        return {"error": f"{e}\n\nTraceback:\n{error_details}"}

class   ChartREPLTool(BaseTool):
    name: str = "custom_python_repl"
    description: str = (
        """A custom REPL tool for executing Python code. Use this to evaluate python scripts and return results.
        """
    )

    def _run(self, code: str) -> str:
        # clean up the code
        code = self.filter_plot_statements(code)
        try:
            execution_result = parse_and_execute_code(code)

            if "error" in execution_result:
                return execution_result["error"]

            # Check if the code includes charting
            if plt.get_fignums() and plt.gca().has_data():
                chart_url = self.save_and_upload_generated_chart()
                # local_namespace["_result"] = chart_url
                return f"Chart uploaded to Azure Blob Storage: {chart_url}"

            return str(execution_result["result"])

        except Exception as e:
            error_details = traceback.format_exc()
            return f"Error: {e}\n\nTraceback:\n{error_details}"

        finally:
            # Ensure the matplotlib state is reset for subsequent calls
            plt.clf()
            plt.close("all")

    async def _arun(self, code: str) -> str:
        raise NotImplementedError("Asynchronous execution is not implemented yet.")

    def filter_plot_statements(self, code: str):
        if "plt.show" in code:
            code = code.replace("plt.show()", "")

        if "plt.savefig" in code:
            code = re.sub(r"plt\.savefig\([^\)]*\)", '""', code)

        if "plt.close" in code:
            code = re.sub(r"plt\.close\([^\)]*\)", '""', code)

        return code

    def save_and_upload_generated_chart(self):
        try:
            chart_path = os.path.join(os.getcwd(), GENERATED_ARTIFACTS_DIRECTORY)
            if not os.path.exists(chart_path):
                os.makedirs(chart_path)  # Create chart directory if it doesn't exist

            # Generate a unique name for the chart
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            unique_id = uuid.uuid4().hex[:8]
            blob_file_name = f"charts/chart_{timestamp}_{unique_id}.png"

            with tempfile.NamedTemporaryFile(suffix=".png", delete=False) as tmp_file:
                chart_path = tmp_file.name
                plt.savefig(chart_path)

            # Upload the chart to Azure Blob Storage
            with open(chart_path, "rb") as file:
                file_data = file.read()
                picture_url = upload_artifact_to_azure(file_data, blob_file_name)

            print(f"Chart uploaded successfully: {picture_url}")
            return picture_url

        except Exception as e:
            error_details = traceback.format_exc()
            print(
                f"Error while saving and uploading chart: {e}\n\nTraceback:\n{error_details}"
            )
            return "Error occurred while saving and uploading the chart."


class ChartToolInput(BaseModel):
    """Input for Chart Tool."""

    data: str = Field(..., description="Data to generate chart from")
    chart_type: str = Field(..., description="Type of chart to generate")


class ChartGenerator(BaseTool):
    name: str = "ChartTool"
    description: str = (
        "This is a charting tool. It Accepts data that is in a JSON string format and then generates a chart based on the data."
    )
    args_schema: Optional[Type[BaseModel]] = ChartToolInput
    return_direct: bool = False

    def _run(self, data: str, chart_type: str) -> str:
        """Execute Python code safely and return the output."""

        prompt = ChatPromptTemplate.from_messages(
            [
                (
                    "system",
                    "You are a Python expert. Write python code to generate charts as required.",
                ),
                ("placeholder", "{chat_history}"),
                ("human", "{input}"),
                ("placeholder", "{agent_scratchpad}"),
            ]
        )
        agent = create_tool_calling_agent(
            llm=LLM,
            tools=[ChartREPLTool()],
            prompt=prompt,
        )
        agent_executor = AgentExecutor(
            agent=agent, tools=[ChartREPLTool()], verbose=True
        )

        result = agent_executor.invoke(
            {"input": f"Generate a {chart_type} chart from the data {data}"}
        )

        return result["output"]

    async def _arun(self, code: str) -> str:
        raise NotImplementedError("Asynchronous execution is not implemented yet.")
```

## Schema

### Agent State

```python
from typing import Any, Dict, List, Optional, TypedDict

class AgentState(TypedDict):
    objective: str
    task_queue: List[Dict]
    completed_tasks: List[Dict]
    failed_tasks: List[Dict]
    current_task: Dict | None
    iteration: int
    max_iterations: int
    task_id_counter: int
    context: Dict[str, Any]
    agent_memory: List[dict]
    validation_approved: bool | None
    validation_feedback: Optional[str]
    validation_attempts: int
    final_response: Optional[str]


class TaskCreation(BaseModel):
    tasks: List[str] = Field(..., description="List of tasks needed to achieve goal")


class ValidationResult(BaseModel):
    valid: bool = Field(..., description="Whether tasks achieve objective")
    feedback: Optional[str] = Field(None, description="Improvement suggestions")


class FinalResponse(BaseModel):
    summary: str = Field(..., description="Comprehensive response to objective")
```

### *Base Prompts*

```python
CREATE_INITIAL_TASK_PROMPT = """
Objective: {objective}

Available tools: {tools}

Validation feedback: {validation_feedback}

NOTE: When generating tasks:
If a question can be responded to immediately, just create a single task that achieves the objective.
ENSURE that none of the tasks are vague.
Take the validation feedback into account.

Generate initial tasks
ensure that the tasks are group under a related heading
display that task list as a todo list using markdown with checkboxes:
"""

VALIDATE_TASK_PROMPT = """
Objective: {objective}
Existing Context: {context}
Available tools: {tools}
Proposed Tasks: {task_descriptions}

Do these tasks fully achieve the objective given current context? Provide feedback.
"""

PRIORITIZE_TASK_PROMPT = """
Objective: {objective}
Tasks to prioritize: {task_descriptions}

Return sorted tasks by priority (most important first).
Add any missing critical tasks.
"""

EXECUTE_TASK_PROMPT = """
Objective: {objective}
Current Task: {description}
"""

HANDLE_FAILURE_PROMPT = """
Objective: {objective}
Failed tasks: {failed_tasks}

Generate sub-tasks or alternative approaches to handle these failures.
"""

COMPILE_RESULT_PROMPT = """
SYNTHESIZE FINAL ANSWER:
Objective: {objective}
Completed Tasks: {completed_tasks}

Combine all task results into a comprehensive response that directly addresses the original objective.
Include supporting details and maintain logical flow.

If any artifacts were generated add them to your final response.

If a summarize exists within the context dont't use it.

NOTE: ALWAYS RETURN A DETAILED AND COMPREHENSIVE RESPONSE.
"""

REACT_AGENT_EXECUTOR_PROMPT = """
You are an AI agent working on: {objective}
Available Tools: {tools}

Current Context:
{context}

Execute the given task
"""
```

## Vanilla Framework

```python
from langchain_core.language_models import BaseLanguageModel
from langchain.tools import BaseTool
from langchain_core.messages import SystemMessage
from langchain_core.runnables.config import RunnableConfig
from langgraph.checkpoint.memory import MemorySaver
from langgraph.constants import END, START
from langgraph.graph import StateGraph
from langgraph.prebuilt import create_react_agent

class LangGraphBabyAGI:
    def __init__(
        self,
        model: BaseLanguageModel,
        tools: List[BaseTool],
        prompt_prefix: str,
        max_iterations: int = 5,
    ):
        self.llm = model
        self.tools = tools
        self.graph = StateGraph(AgentState)
        self.executor_agent = None
        self.tool_dict = {t.name: t for t in tools}
        self.builder = StateGraph(AgentState)
        self.max_iterations = max_iterations
        self.prompt_prefix = prompt_prefix
        self._build_graph()

        # GRAPH BUILDER

    def to_dict(self) -> dict:
        """Serialize the agent's configuration for JSON serialization."""
        return {
            "prompt_prefix": self.prompt_prefix,
            "max_iterations": self.max_iterations,
            "tools": [tool.name for tool in self.tools],
        }

    def _build_graph(self):
        # Add Nodes
        self.builder.add_node("create_initial_tasks", self._create_initial_tasks)
        self.builder.add_node("validate_tasks", self._validate_tasks)
        self.builder.add_node("prioritize_tasks", self._prioritize_tasks)
        self.builder.add_node("execute_task", self._execute_task)
        self.builder.add_node("handle_failure", self._handle_failure)
        self.builder.add_node("compile_results", self._compile_results)

        # Add edges
        self.builder.add_edge(START, "create_initial_tasks")
        self.builder.add_edge("create_initial_tasks", "validate_tasks")
        self.builder.add_conditional_edges(
            "validate_tasks",
            self._decide_validation,
            {
                "valid": "prioritize_tasks",
                "invalid": "create_initial_tasks",
                "max_exceeded": "compile_results",
            },
        )
        self.builder.add_conditional_edges(
            "execute_task",
            self._decide_after_execution,
            {
                "continue": "execute_task",
                "fail": "handle_failure",
                "complete": "compile_results",
            },
        )
        self.builder.add_edge("handle_failure", "prioritize_tasks")
        self.builder.add_conditional_edges(
            "prioritize_tasks",
            self._check_iteration_limit,
            {"continue": "execute_task", "stop": "compile_results"},
        )
        self.builder.add_edge("compile_results", END)

        memory = MemorySaver()
        self.graph = self.builder.compile(checkpointer=memory)
        return

    def _create_initial_tasks(self, state: AgentState) -> Dict:
        # Format tools as a comma-separated string for clarity.
        print("CURRENTLY IN CREATE INITIAL TASKS STAGE")
        tools_str = ", ".join(self.tool_dict.keys())

        create_initial_tasks_prompt = (
            self.prompt_prefix
            + CREATE_INITIAL_TASK_PROMPT.format(
                objective=state["objective"],
                tools=tools_str,
                validation_feedback=state.get("validation_feedback", "None"),
            )
        )

        structured_llm = self.llm.with_structured_output(TaskCreation)

        response = structured_llm.invoke(
            [SystemMessage(content=create_initial_tasks_prompt)]
        )

        tasks = [
            {
                "task_id": state["task_id_counter"] + i,
                "description": task,
                "attempts": 0,
                "status": "pending",
            }
            for i, task in enumerate(response.tasks, 1)
        ]

        return {
            "task_queue": tasks,
            "task_id_counter": state["task_id_counter"] + len(tasks),
            "validation_approved": False,
        }

    def _validate_tasks(self, state: AgentState) -> Dict:
        print("CURRENTLY IN VALIDATE TASKS STAGE")

        task_descriptions = [t["description"] for t in state["task_queue"]]

        task_validation_prompt = self.prompt_prefix + VALIDATE_TASK_PROMPT.format(
            objective=state["objective"],
            context=self._format_context(state["context"]),
            task_descriptions=task_descriptions,
            tools=self.tool_dict.keys(),
        )

        structured_llm = self.llm.with_structured_output(ValidationResult)

        new_attempts = state["validation_attempts"]

        response = structured_llm.invoke(
            [SystemMessage(content=task_validation_prompt)]
        )

        new_attempts += 1

        return {
            "validation_approved": response.valid,
            "validation_feedback": response.feedback,
            "validation_attempts": new_attempts,
        }

    def _prioritize_tasks(self, state: AgentState) -> Dict:
        print("CURRENTLY IN PRIORITIZE STAGE")
        structured_llm = self.llm.with_structured_output(TaskCreation)

        all_tasks = state["task_queue"] + state["failed_tasks"]
        task_descriptions = [t["description"] for t in all_tasks]
        iteration = state["iteration"]

        prioritize_task_prompt = PRIORITIZE_TASK_PROMPT.format(
            objective=state["objective"],
            task_descriptions=task_descriptions,
        )

        response = structured_llm.invoke(
            [SystemMessage(content=prioritize_task_prompt)]
        )

        task_map = {t["description"]: t for t in all_tasks}
        prioritized = []
        for desc in response.tasks:
            if desc in task_map:
                prioritized.append(task_map[desc])
            else:
                prioritized.append(
                    {
                        "task_id": state["task_id_counter"],
                        "description": desc,
                        "attempts": 0,
                        "status": "pending",
                    }
                )
                state["task_id_counter"] += 1

        return {
            "task_queue": prioritized,
            "failed_tasks": [],
            "iteration": iteration + 1,
        }

    def _execute_task(self, state: AgentState) -> Dict:
        print("CURRENTLY IN EXECUTE STAGE")
        agent_executor = self._get_agent(state)

        if not state["task_queue"]:
            return {}

        task = state["task_queue"].pop(0)
        state["current_task"] = task
        task["attempts"] += 1

        try:
            # Prepare agent input with context
            task_prompt = EXECUTE_TASK_PROMPT.format(
                objective=state["objective"], description=task["description"]
            )

            # Execute the react agent
            result = agent_executor.invoke(
                ({"messages": task_prompt}),
                config={
                    "configurable": {"thread_id": state["objective"]},
                },
            )

            return self._finalize_task(state, result["messages"][-1].content)

        except Exception as e:
            error_msg = f"{type(e).__name__}: {str(e)}"
            return {
                "current_task": {**task, "status": "failed", "error": error_msg},
                "failed_tasks": state["failed_tasks"] + [task],
            }

    def _handle_failure(self, state: AgentState) -> Dict:
        print("CURRENTLY IN HANDLE FAILURE STAGE")
        structured_llm = self.llm.with_structured_output(TaskCreation)

        failed_tasks = [t["description"] for t in state.get("failed_tasks")]

        handle_failure_prompt = HANDLE_FAILURE_PROMPT.format(
            objective=state["objective"], failed_tasks=failed_tasks
        )

        response = structured_llm.invoke([SystemMessage(content=handle_failure_prompt)])

        new_tasks = [
            {
                "task_id": state["task_id_counter"] + i,
                "description": task,
                "attempts": 0,
                "status": "pending",
            }
            for i, task in enumerate(response.tasks, 1)
        ]

        return {
            "task_queue": state["task_queue"] + new_tasks,
            "task_id_counter": state["task_id_counter"] + len(new_tasks),
            "failed_tasks": [],
        }

    def _compile_results(self, state: AgentState) -> Dict:
        structured_llm = self.llm.with_structured_output(FinalResponse)

        completed = "\n".join(
            f"- {task['description']}\n  Result: {task.get('result', 'No result')}"
            for task in state["completed_tasks"]
        )

        compile_prompt = COMPILE_RESULT_PROMPT.format(
            objective=state["objective"], completed_tasks=completed
        )

        response = structured_llm.invoke([SystemMessage(content=compile_prompt)])

        return {"final_response": response.summary}

    # UTILITIES
    def _finalize_task(self, state: AgentState, answer: str) -> Dict:
        task = state["current_task"]
        context_key = f"task_{task['task_id']}"
        state["context"][context_key] = {
            "description": task["description"],
            "result": answer,
        }

        return {
            "completed_tasks": state["completed_tasks"]
            + [{**task, "result": answer, "status": "completed"}],
            "current_task": task,
        }

    def _get_agent(self, state: AgentState):
        system_prompt = REACT_AGENT_EXECUTOR_PROMPT

        if not self.executor_agent:
            memory = MemorySaver()
            self.executor_agent = create_react_agent(
                model=self.llm,
                tools=self.tools,
                prompt=system_prompt.format(
                    objective=state["objective"],
                    tools=self.tool_dict.keys(),
                    context=self._format_context(state["context"]),
                ),
                name="react",
                checkpointer=memory,
            )
            return self.executor_agent
        return self.executor_agent

    def _format_context(self, context: Dict) -> str:
        if not context:
            return "No context available yet"
        return "\n".join(
            f"- {key}: {value['description']}\n  Result: {value['result'][:200]}..."
            for key, value in context.items()
        )

    # Decision makers
    def _decide_validation(self, state: AgentState) -> str:
        if state["validation_approved"]:
            return "valid"
        elif state["validation_attempts"] >= state["max_iterations"]:
            return "max_exceeded"
        return "invalid"

    def _check_iteration_limit(self, state: AgentState) -> str:
        return "stop" if state["iteration"] >= state["max_iterations"] else "continue"

    def _decide_after_execution(self, state: AgentState) -> str:
        if not state["task_queue"] and not state["failed_tasks"]:
            return "complete"
        if state.get("current_task", {}).get("status") == "failed":
            return "fail"
        return "continue"

    def invoke(self, objective: str, config: Optional[RunnableConfig]) -> Dict:
        """Execute the agent workflow and return final state"""
        initial_state = AgentState(
            objective=objective,
            task_queue=[],
            completed_tasks=[],
            failed_tasks=[],
            current_task=None,
            iteration=0,
            max_iterations=self.max_iterations,
            task_id_counter=0,
            context={},
            agent_memory=[],
            validation_approved=None,
            validation_feedback=None,
            validation_attempts=0,
            final_response=None,
        )

        # self.graph.invoke(initial_state, config)
        for event in self.graph.stream(initial_state, config, stream_mode="values"):
            tasks = event.get("task_queue", None)
            validation = event.get("validation_approved", None)
            validation_feedback = event.get("validation_feedback", None)
            iteration = event.get("iteration", None)
            print("ITERATION!!!!", iteration)

            if validation is not None:
                print(f"Validation Approved: {validation}")

            print(f"validation feeback:{validation_feedback}")

            if tasks:
                for task in tasks:
                    print(f"Task ID: {task['task_id']} : {task['description']}")

        print(
            f"FINAL_RESPONSE: {self.graph.get_state(config).values.get('final_response')}"
        )
        return self.graph.get_state(config)
```

## Expert Swarm Framework

### Expert Agent

```python
from typing import Any

from langchain_core.language_models import BaseLanguageModel
from pydantic import BaseModel, Field


class ExpertResponse(BaseModel):
    expert_name: str = Field(..., description="Name of the expert")
    contribution: str = Field(..., description="The contribution of the expert")


class ExpertAgent:
    def __init__(self, role: str, llm: BaseLanguageModel, job_description: str):
        self.role = role
        self.llm = llm
        self.job_description = job_description

    def __repr__(self):
        return f"ExpertAgent(role='{self.role}', job_description='{self.job_description[:50]}...')"

    def contribute(self, context: str) -> ExpertResponse:
        prompt: str = (
            f"Your name is {self.role}. You are an expert in {self.role} with a job description of {self.job_description}. Given the following context, "
            "do the task/objective you were asked to in the context:\n\n"
            f"{context}\n\n"
            "Your response:"
        )

        structured_llm = self.llm.with_structured_output(ExpertResponse)

        response: ExpertResponse | Any = structured_llm.invoke(
            [SystemMessage(content=prompt)]
        )
        return response
```

### Generate Experts

```python
from typing import List

from langchain_core.language_models import BaseLanguageModel
from pydantic import BaseModel, Field


class GeneratedExpert(BaseModel):
    role: str = Field(..., description="Role of expert relevant to the objective")
    job_description: str = Field(
        ..., description="Job description of expert relevant to the objective"
    )


class GeneratedExperts(BaseModel):
    experts: List[GeneratedExpert] = Field(
        ..., description="List of experts relevant to the objective"
    )


# The power is in the prompt here. Currently I have set it to return the minimum most efficient number of agents possible but I have tried to force it to create max_experts agents too and that works fine albeit more resource consuming
def generate_experts(
    model: BaseLanguageModel, objective: str, max_experts: int = 4, min_experts: int = 1
) -> List[ExpertAgent]:
    # Create a prompt that instructs the LLM to generate a list of experts.
    prompt_text = f"""
    Identify the most suitable expert agents for the given objective.

    Objective: {objective}

    Select only the experts whose knowledge and skills are most relevant to achieving the objective.
    - The maximum number of experts is {max_experts}, but choose the **minimum** required based on task complexity.
    - Return at least **{min_experts} experts**, but avoid unnecessary roles.

    """
    # Creating a new model without cache
    model = LLM
    structured_llm = model.with_structured_output(GeneratedExperts)
    response = structured_llm.invoke(prompt_text)

    # no_cache_llm = structured_llm.configurable_alternatives(
    #     ConfigurableField(id="cache_enabled"),
    #     default_key="disabled",
    #     disabled={"cache": None}
    # )

    expert_agents = [
        ExpertAgent(role=agent.role, llm=model, job_description=agent.job_description)
        for agent in response.experts
    ]

    return expert_agents
```

### Multi Expert Collaborator

```python
class MultiExpertCollaborator:
    def __init__(self, experts: list[ExpertAgent]):
        self.experts = experts

    def aggregate_contributions(self, contributions: List[ExpertResponse]) -> str:
        aggregated = "\n\n".join(
            [f"{con.expert_name} says:\n{con.contribution}" for con in contributions]
        )
        return aggregated

    def collaborate(self, context: str) -> str:
        contributions = []
        for expert in self.experts:
            contribution = expert.contribute(context)
            contributions.append(contribution)
        final_response = self.aggregate_contributions(contributions)
        return final_response

    def synthesize_response(
        self,
        initial_prompt: str,
        structured_llm: Any,
        meta_prompt: str = "Based on the above expert insights, generate an optimized final response. Ensure that you return it in the exact format that each of them handed it off to you.",
        name: str = "Lola",
    ) -> Any:
        meta_template = """
        Your are an aggregator and synthesizing expert. You are never needed to provide new information, just to aggregate and synthesize information from other experts
                Aggregated Expert Input:
        {aggregated_expert_response}

        """
        meta_template += meta_prompt

        aggregated_expert_response = self.collaborate(initial_prompt)
        print("*" * 10)
        print(aggregated_expert_response)
        print("*" * 10)

        meta_prompt = meta_template.format(
            aggregated_expert_response=aggregated_expert_response
        )

        response = structured_llm.invoke([SystemMessage(content=meta_prompt)])
        return response
```

### Expert Swarm

```python
class ExpertSwarm(LangGraphBabyAGI):
    """
    An expert swarm variant of LangGraphBabyAGI that uses multiple expert insights
    to create an optimized list of initial tasks.
    """

    def __init__(
        self,
        model: BaseLanguageModel,
        tools: List[BaseTool],
        prompt_prefix: str,
        max_iterations: int = 5,
        is_expert_list_preset: bool = True,
        min_experts: int = 4,
        max_experts: int = 4,
    ):
        super().__init__(model, tools, prompt_prefix, max_iterations)
        self.experts = List[ExpertAgent]
        self.is_expert_list_preset = is_expert_list_preset
        self.min_experts = min_experts
        self.max_experts = min_experts if min_experts > max_experts else max_experts

    def _create_initial_tasks(self, state: AgentState) -> Dict:
        structured_llm = self.llm.with_structured_output(TaskCreation)

        tools_str = ", ".join(self.tool_dict.keys())
        validation_feedback = state.get("validation_feedback", "None")

        create_initial_tasks_prompt = CREATE_INITIAL_TASK_PROMPT.format(
            objective=state["objective"],
            tools=tools_str,
            validation_feedback=validation_feedback,
        )
        # meta_prompt = """Based on the above expert insights, generate an optimized list of tasks to achieve the objective.
        # Select only the best, most actionable tasks."""

        experts = generate_experts(
            self.llm,
            state["objective"],
            min_experts=self.min_experts,
            max_experts=self.max_experts,
        )
        self.experts = experts

        multi_expert = MultiExpertCollaborator(experts=experts)
        response = multi_expert.synthesize_response(
            initial_prompt=create_initial_tasks_prompt,
            structured_llm=structured_llm,
            meta_prompt=self.prompt_prefix,
        )

        tasks = [
            {
                "task_id": state["task_id_counter"] + i,
                "description": task,
                "attempts": 0,
                "status": "pending",
            }
            for i, task in enumerate(response.tasks, start=1)
        ]

        return {
            "task_queue": tasks,
            "task_id_counter": state["task_id_counter"] + len(tasks),
            "validation_approved": False,
        }

    def _validate_tasks(self, state: AgentState) -> Dict:
        task_descriptions = [t["description"] for t in state["task_queue"]]

        structured_llm = self.llm.with_structured_output(ValidationResult)

        new_attempts = state["validation_attempts"]

        # meta_prompt = """Based on the above expert contributions, synthesize a final response that integrates their diverse insights into one coherent and actionable summary. Return it in the format that each of them presented it.
        # """

        task_validation_prompt = VALIDATE_TASK_PROMPT.format(
            objective=state["objective"],
            context=self._format_context(state["context"]),
            task_descriptions=task_descriptions,
            tools=self.tool_dict.keys(),
        )

        experts = (
            self.experts
            if self.is_expert_list_preset
            else generate_experts(
                self.llm,
                state["objective"],
                min_experts=self.min_experts,
                max_experts=self.max_experts,
            )
        )

        multi_expert = MultiExpertCollaborator(experts=experts)
        response = multi_expert.synthesize_response(
            initial_prompt=task_validation_prompt, structured_llm=structured_llm
        )

        new_attempts += 1

        return {
            "validation_approved": response.valid,
            "validation_feedback": response.feedback,
            "validation_attempts": new_attempts,
        }

    def _prioritize_tasks(self, state: AgentState) -> Dict:
        structured_llm = self.llm.with_structured_output(TaskCreation)

        all_tasks = state["task_queue"] + state["failed_tasks"]
        task_descriptions = [t["description"] for t in all_tasks]
        iteration = state["iteration"]

        meta_prompt = """Your task is not to aggregate new ideas but to review the priority order and votes provided by the experts. Analyze the expert suggestions and their associated votes, then produce a final, prioritized list. Ensure the final list is sorted so that the items with the highest consensus and votes appear first.

        """

        prioritize_task_prompt = PRIORITIZE_TASK_PROMPT.format(
            objective=state["objective"],
            task_descriptions=task_descriptions,
        )

        experts = (
            self.experts
            if self.is_expert_list_preset
            else generate_experts(
                self.llm,
                state["objective"],
                min_experts=self.min_experts,
                max_experts=self.max_experts,
            )
        )

        multi_expert = MultiExpertCollaborator(experts=experts)
        response = multi_expert.synthesize_response(
            initial_prompt=prioritize_task_prompt,
            structured_llm=structured_llm,
            meta_prompt=meta_prompt,
        )

        task_map = {t["description"]: t for t in all_tasks}
        prioritized = []
        for desc in response.tasks:
            if desc in task_map:
                prioritized.append(task_map[desc])
            else:
                prioritized.append(
                    {
                        "task_id": state["task_id_counter"],
                        "description": desc,
                        "attempts": 0,
                        "status": "pending",
                    }
                )
                state["task_id_counter"] += 1

        return {
            "task_queue": prioritized,
            "failed_tasks": [],
            "iteration": iteration + 1,
        }

    def _handle_failure(self, state: AgentState) -> Dict:
        structured_llm = self.llm.with_structured_output(TaskCreation)

        failed_tasks = [t["description"] for t in state["failed_tasks"]]

        # meta_prompt = """Based on the above expert contributions, synthesize a final response that integrates their diverse insights into one coherent and actionable summary. Return it in the format that each of them presented it.
        # """

        handle_failure_prompt = HANDLE_FAILURE_PROMPT.format(
            objective=state["objective"], failed_tasks=failed_tasks
        )

        experts = (
            self.experts
            if self.is_expert_list_preset
            else generate_experts(
                self.llm,
                state["objective"],
                min_experts=self.min_experts,
                max_experts=self.max_experts,
            )
        )

        multi_expert = MultiExpertCollaborator(experts=experts)
        response = multi_expert.synthesize_response(
            initial_prompt=handle_failure_prompt, structured_llm=structured_llm
        )

        new_tasks = [
            {
                "task_id": state["task_id_counter"] + i,
                "description": task,
                "attempts": 0,
                "status": "pending",
            }
            for i, task in enumerate(response.tasks, 1)
        ]

        return {
            "task_queue": state["task_queue"] + new_tasks,
            "task_id_counter": state["task_id_counter"] + len(new_tasks),
            "failed_tasks": [],
        }
```

```python
from uuid import uuid4
from collections.abc import Iterable
import json
from langchain.agents import AgentExecutor
from langchain.evaluation import load_evaluator
from typing import Literal

def run_benchmark(dataset: Iterable, agent_engine: AgentExecutor, thread_id = f"{uuid4()}", file_save="agent_gaia_result", start=0, stop=None):
  stop = len(dataset) if stop is None else stop
  dataset = dataset.select(range(start, stop))
  thread = {"configurable": {"thread_id": thread_id, "recursion_limit": 100}, "recursion_limit": 100}

  answer = {}
  predictions = {}
  data = {}

  for i, item in enumerate(dataset, start):
    print(f"🚀 Now processing: ⭐ Benchmark Sample {i+1}/{stop} ⭐")
    answer[i] = item["Final answer"]
    snapshot = agent_engine.invoke(
            item["Question"], thread
        )
    pred = snapshot.values["final_response"]
    predictions[i] = pred
    data[i] = {"question": item["Question"],
               "agent_pred":pred,
               "truth": item["truth"],
               "Level": item["Level"],
               "file_name'": item["file_name'"],
               "file_path": item["file_path"],
               "Annotator Metadata": item["Annotator Metadata"],
               }

    if((i-start)%5 == 0):
        accuracy, updated_data = evaluate_exact_match(data)
        save_dict_to_file(data=updated_data, accuracy=accuracy, file_name=f"checkpoint_{file_save}")

  accuracy, updated_data = evaluate_exact_match(data)
  save_dict_to_file(data=updated_data, accuracy=accuracy, file_name=file_save)

  return accuracy, data, snapshot

from langchain.evaluation import load_evaluator

def evaluate_exact_match(data: dict):
  if not data:
      return 0.0, {}  # Return 0% and empty results if no data is provided

  score = 0
  total = len(data)
  evaluation_results = {}

  evaluator = load_evaluator("qa", llm=LLM)  # Load evaluator once to optimize

  for key, value in data.items():
    eval_result = evaluator.evaluate_strings(
        prediction=value["pred"],
        input=value["question"],
        reference=value["truth"],
    )

    is_correct = int(eval_result["score"]) > 0
    score += int(eval_result["score"])  # Increment score based on evaluation

    # Store the evaluation result with correctness flag
    evaluation_results[key] = {
        "question": value["question"],
        "pred": value["pred"],
        "truth": value["truth"],
        # "score": int(eval_result["score"]),
        "correct": is_correct,
    }

  accuracy = (score / total) * 100
  print(f"\nAccuracy is {accuracy:.2f}%\n")

  return accuracy, evaluation_results


def save_dict_to_file(data: dict, accuracy: float, file_name: str, upload_to:Literal["azure","drive"]= "drive" ):
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    file_path = f"{file_name}_{timestamp}.json"
    # Add accuracy to the dictionary
    data["accuracy"] = accuracy

    # Save everything in a single JSON object
    with open(file_path, "w") as file:
        json.dump(data, file, indent=4)

    if upload_to == "azure":
      file_bytes=json.dumps(data).encode('utf-8')
      # with open(file_path, "rb") as file:
      #   file_bytes = file.read()
      upload_artifact_to_azure(file_data=file_bytes, file_name=f"gaia_benchmarks/{file_path}")
    if upload_to == "drive":
      !cp $file_path "/content/drive/MyDrive/Eagle Eye/Engineering /gaia_results"

def load_file_and_evaluate(file_path: str, file_save = "final_result"):
  with open(file_path, "r") as file:
    content = file.read()

  dict_content = json.loads(content)
  accuracy, updated_data = evaluate_exact_match(dict_content)
  save_dict_to_file(data=updated_data, accuracy=accuracy, file_name=file_save)
```
